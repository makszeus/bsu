```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# 
d = 1.0  #   
N = 1000  #   
x = np.linspace(0, 2*d, N)  # 
h = x[1] - x[0]  #  

#  V(x) = x^2
V = (x - d)**2

#   
T = -2*np.eye(N) + np.eye(N, k=1) + np.eye(N, k=-1)
T = -T / (h**2)

#   
H = T + np.diag(V)

#     
eigvals, eigvecs = la.eigh(H)

#         
for i in range(3):
    print(f'Eigenvalue {i}: {eigvals[i]}')
    plt.plot(x, eigvecs[:, i], label=f'{i} (E={eigvals[i]:.2f})')

plt.xlabel('x')
plt.ylabel('(x)')
plt.legend()
plt.title('Eigenfunctions of the Schrdinger equation')
plt.grid()
plt.show()
```
