```python
# Task 12. Variant 9.

import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# 
d = 1.0  #   
N = 1000  #   
x = np.linspace(0, 2*d, N)  # 
h = x[1] - x[0]  #  

#  V(x) = (x - d)^2
V = (x - d)**2

#   
hbar = 1.0  #   
m = 1.0  #  

#      
dt = 0.001

#  :      t=0
Psi = np.zeros(N, dtype=complex)
Psi[0] = 0
Psi[-1] = 0

#        ""
for t in range(1000):
    #       x
    d2Psi_dx2 = (Psi[2:] - 2*Psi[1:-1] + Psi[:-2]) / (h**2)
    
    #       
    Psi[1:-1] += (-1j*hbar/(2*m) * d2Psi_dx2 + 1j*V[1:-1]*Psi[1:-1]) * dt

#       
eigvals, eigvecs = la.eigh(np.diag(V) - (hbar**2 / (2*m*h**2)) * (np.diag(np.ones(N-1), 1) - 2*np.diag(np.ones(N), 0) + np.diag(np.ones(N-1), -1)))

#         
for i in range(3):
    print(f'Eigenvalue {i}: {eigvals[i]}')
    plt.plot(x, eigvecs[:, i], label=f'{i} (E={eigvals[i]:.2f})')

plt.xlabel('x')
plt.ylabel('(x)')
plt.legend()
plt.title('Eigenfunctions of the Schrdinger equation')
plt.grid()
plt.show()
```
