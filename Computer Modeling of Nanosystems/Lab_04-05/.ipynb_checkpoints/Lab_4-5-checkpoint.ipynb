```python
# Task 4-5. Variant 9.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize


def func_1(x):
    return 2 * x[0] ** 3 - x[0] * x[1] ** 2 + 11 * x[0] ** 2 + 16 * x[0] + 7

def plot_func_1():
    x_values = np.linspace(-2, 2)
    y_values = np.linspace(-2, 2)
    
    xg, yg = np.meshgrid(x_values, y_values)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(xg, yg, func_1([xg, yg]), rstride=1, cstride=1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x, y)')
    ax.set_title('$f(x,y)=2 * x^3 - xy^2 +11 * x^2 + 16 * x + 7$')

def func_2(x):
    return 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2

def plot_func_2():
    x_values = np.linspace(-2, 2)
    y_values = np.linspace(-2, 2)
    
    xg, yg = np.meshgrid(x_values, y_values)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(xg, yg, func_2([xg, yg]), rstride=1, cstride=1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x, y)')
    ax.set_title('$f(x_1,x_2)=100(x_2-x_1^2)^2+(1-x_1)^2$')

if __name__ == "__main__":
    plot_func_1()
    plot_func_2()
```

```python
import numpy as np

# Определение метода Хука-Дживса для минимизации функции
def hook_jeeves_method(func, x0, step, eps):
    iter_counter = 0  # Инициализация счётчика итераций
    n = len(x0)  # Количество переменных (размерность задачи)
    b = np.array(x0)  # Начальная точка
    h = np.array([step] * n)  # Шаг поиска для каждой переменной
    x_history = [b.copy()]  # История точек, посещённых алгоритмом

    # Вспомогательная функция для оценки значения функции
    def evaluate(x):
        return func(x)

    # Поиск по образцу (pattern search) вокруг текущей точки x с шагом h
    def pattern_search(x, h):
        f_best = evaluate(x)
        x_best = x.copy()

        for i in range(n):
            # Увеличение x_i
            x_new = x.copy()
            x_new[i] += h[i]
            f_new = evaluate(x_new)
            if f_new < f_best:
                f_best = f_new
                x_best = x_new

            # Уменьшение x_i
            x_new = x.copy()
            x_new[i] -= h[i]
            f_new = evaluate(x_new)
            if f_new < f_best:
                f_best = f_new
                x_best = x_new

        return x_best, f_best

    # Основной цикл метода Хука-Дживса
    while np.any(h > eps):
        iter_counter += 1
        x_prev = b.copy()

        # Выполнение поиска по образцу
        b, f_b = pattern_search(b, h)
        x_history.append(b.copy())

        # Если текущая точка b осталась неизменной, то уменьшаем шаг поиска
        if np.allclose(b, x_prev):
            h /= 10
        # Иначе выполняем исследовательский поиск
        else:
            x_next, f_next = pattern_search(b + 2 * (b - x_prev), h)
            if f_next < f_b:
                b = x_next
                x_history.append(b.copy())

    return x_history, iter_counter

# Функция для визуализации итераций метода Хука-Дживса на контурной карте функции
def plot_iterations(x_history, f):
    x_history = np.array(x_history)
    x = np.linspace(-5, 5, 400)
    y = np.linspace(-5, 5, 400)
    X, Y = np.meshgrid(x, y)
    Z = f([X, Y])
    
    plt.figure(figsize=(8, 6))
    
    # Построение контуров уровней функции
    plt.contour(X, Y, Z, levels=np.logspace(0, 5, 35), cmap='jet')
    # Нанесение точек из истории итераций
    plt.plot(x_history[:, 0], x_history[:, 1], 'ko-', markersize=4, linewidth=0.8)
    plt.colorbar()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Итерации метода Хука-Дживса')
    plt.show()

if __name__ == "__main__":
    x0_1 = [1.0, 1.0]
    x0_2 = [-1.0, 1.0]
    step = 1.0
    eps = 1e-6
    
    # Применение метода Хука-Дживса для первой функции и начальной точки
    x_history_1, iter_counter_1 = hook_jeeves_method(func_1, x0_1, step, eps)
    print(f"Результат для func_1: {x_history_1[-1]}, количество итераций: {iter_counter_1}")
    plot_iterations(x_history_1, func_1)
    
    # Применение метода Хука-Дживса для второй функции и начальной точки
    x_history_2, iter_counter_2 = result_2 = hook_jeeves_method(func_2, x0_2, step, eps)
    print(f"Результат для func_2: {x_history_2[-1]}, количество итераций: {iter_counter_2}")
    plot_iterations(x_history_2, func_2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import approx_fprime
from scipy.linalg import inv

# Тестовые функции
def func_1(x):
    return 2 * x[0] ** 3 - x[0] * x[1] ** 2 + 11 * x[0] ** 2 + 16 * x[0] + 7

def func_2(x):
    return 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2

# Приближённое численное вычисление градиента с помощью функции scipy
def grad_func_1(x):
    return approx_fprime(x, func_1, epsilon=1e-6)

def grad_func_2(x):
    return approx_fprime(x, func_2, epsilon=1e-6)

# Явное вычисление гессиана (вторая производная) для func_1
def hessian_func_1(x):
    hessian = np.zeros((2, 2))
    hessian[0, 0] = 12 * x[0] + 22
    hessian[0, 1] = -2 * x[1]
    hessian[1, 0] = hessian[0, 1]
    hessian[1, 1] = -2 * x[0]
    return hessian

# Явное вычисление гессиана для func_2
def hessian_func_2(x):
    hessian = np.zeros((2, 2))
    hessian[0, 0] = 2 + 1200 * x[0]**2 - 400 * x[1]
    hessian[0, 1] = -400 * x[0]
    hessian[1, 0] = hessian[0, 1]
    hessian[1, 1] = 200
    return hessian

# Реализация метода Марквардта
def markwardt_method(f, grad, hessian, x0, eps, M, m_0):
    iter_counter = 0  # Счётчик итераций
    x = x0  # Текущая точка
    k = 0  # Счётчик итераций Марквардта
    m_k = m_0  # Параметр m_k, начальное значение
    x_history = [x]  # История итераций
    
    while True:
        iter_counter += 1
        grad_f = grad(x)  # Вычисление градиента в текущей точке
        
        # Проверка условий остановки
        if np.linalg.norm(grad_f) < eps or k >= M:
            return x_history, iter_counter
        
        # Вычисление скорректированного гессиана
        H_m = hessian(x) + m_k * np.eye(len(x))
        
        # Обновление точки x
        x_next = x - np.dot(inv(H_m), grad_f)
        f_x_next = f(x_next)
        
        # Если значение функции уменьшилось, уменьшаем m_k
        if f_x_next < f(x):
            k += 1
            m_k /= 2
            x = x_next
        else:
            m_k *= 2  # Иначе увеличиваем m_k
            
        x_history.append(x)  # Сохраняем точку в историю

# Функция для визуализации итераций
def plot_iterations(x_history, f):
    x_history = np.array(x_history)
    x = np.linspace(-5, 5, 400)
    y = np.linspace(-5, 5, 400)
    X, Y = np.meshgrid(x, y)
    Z = f([X, Y])
    
    plt.figure(figsize=(8, 6))
    plt.contour(X, Y, Z, levels=np.logspace(0, 5, 35), cmap='jet')
    plt.plot(x_history[:, 0], x_history[:, 1], 'ko-', markersize=4, linewidth=0.8)
    plt.colorbar()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Итерации метода Марквардта')
    plt.show()

# Основная часть программы
if __name__ == "__main__":
    x0 = np.array([1, 1])  # Начальное приближение
    eps = 1e-6  # Порог для нормы градиента
    M = 100  # Предельное число итераций
    m_0 = 1  # Начальное значение параметра m_k

    # Оптимизация для func_1
    x_history_1, iter_counter_1 = markwardt_method(func_1, grad_func_1, hessian_func_1, x0, eps, M, m_0)
    print(f"Результат для func_1: {x_history_1[-1]}, количество итераций: {iter_counter_1}")
    plot_iterations(x_history_1, func_1)
    
    x0 = np.array([-1, 1])  # Начальное приближение
    eps = 1e-6  # Порог для нормы градиента
    M = 100  # Предельное число итераций
    m_0 = 1  # Начальное значение параметра m_k

    # Оптимизация для func_2
    x_history_2, iter_counter_2 = markwardt_method(func_2, grad_func_2, hessian_func_2, x0, eps, M, m_0)
    print(f"Результат для func_2: {x_history_2[-1]}, количество итераций: {iter_counter_2}")
    plot_iterations(x_history_2, func_2)
```
