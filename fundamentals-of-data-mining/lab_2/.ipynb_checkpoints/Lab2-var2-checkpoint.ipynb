```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# task 1 prep

df_train_all = pd.read_csv('train.csv')
df_train_all.head()
```

```python
plt.hist(df_train_all['LotArea'], bins=100, color='blue', alpha=0.7)

# Настройка осей и заголовка
plt.xlabel('<LotArea> значения')
plt.ylabel('Частота')
plt.title('Гистограмма LotArea')

# Отображение гистограммы
plt.show()

# Похоже на нормальное, вроде...
```

```python
df_train = df_train_all['LotArea']
threshold = 3
z_scores = np.abs((df_train - df_train.mean()) / df_train.std())
outliers = df_train[z_scores > threshold]
print("Аномалии: \n", outliers)
print("Среднее абсолютное аномалий:", outliers.abs().mean())
print("Минимальное абсолютное аномалий:", outliers.abs().min())
```

```python
Q1 = df_train.quantile(0.25)
Q3 = df_train.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_train[(df_train < lower_bound) | (df_train > upper_bound)]

print("Аномалии: \n", outliers)
print("Среднее абсолютное аномалий:", outliers.abs().mean())
print("Минимальное абсолютное аномалий:", outliers.abs().min())
```

```python
# task 1 choice

df_train_filt_z = df_train[z_scores <= threshold]
df_train_filt_iqr = df_train[(df_train >= lower_bound) & (df_train <= upper_bound)]
print(df_train_filt_z)
print(df_train_filt_iqr)
print('Median: ', df_train.median())

# Всё же выберем фильтрацию по z-score 
```

```python
# task 2, task 3

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = df_train_filt_z

# Построение гистограммы
count, bins, ignored = plt.hist(data, bins=100, density=True, alpha=0.6, color='g')

# Построение кривой нормального распределения
mu, sigma = np.mean(data), np.std(data)
best_fit_line = stats.norm.pdf(bins, mu, sigma)
plt.plot(bins, best_fit_line, 'r-', linewidth=2)

# Оценка параметров распределения Коши
x0 = np.median(data)
gamma = np.median(np.abs(data - x0))
# Построение кривой распределения Коши
best_fit_line = stats.cauchy.pdf(bins, loc=x0, scale=gamma)
plt.plot(bins, best_fit_line, 'y-', linewidth=2)

# Вычисление параметров гамма-распределения
k = (mu ** 2) / (sigma ** 2)
theta = (sigma ** 2) / mu
# Построение кривой гамма-распределения
best_fit_line = stats.gamma.pdf(bins, k, scale=theta)
plt.plot(bins, best_fit_line, 'b-', linewidth=2)

plt.title("Гистограмма с наложенными кривыми непрерывных распределений")
plt.xlabel("Значение")
plt.ylabel("Плотность вероятности")
plt.show()

# Данные сгенерированы из распределения Коши, да?
```

```python
# task 4

import numpy as np
import scipy.stats as stats

# used data - 'data' array from task 3
expected_distribution = stats.cauchy.rvs(loc=x0, scale=gamma, size=len(data))

# Построение гистограммы для наблюдаемых данных
observed_freq, observed_bins = np.histogram(data, bins=15)
# Построение гистограммы для ожидаемых данных
expected_freq, expected_bins = np.histogram(expected_distribution, bins=observed_bins)

# Нормализация частот
observed_freq = observed_freq / observed_freq.sum()
expected_freq = expected_freq / expected_freq.sum()

# Проведение хи-квадрат теста (самодельный)
'''
Нулевая гипотеза (H₀): данные соответствуют ожидаемому распределению (расп. Коши).
Альтернативная гипотеза (H₁): данные не соответствуют ожидаемому распределению.
'''
# Вычисление статистики хи-квадрат вручную
chi2_stat = np.sum((observed_freq - expected_freq)**2 / expected_freq)
print(f"Хи-квадрат статистика: {chi2_stat}")

# Определение критического значения и p-значения
alpha = 0.05  # Уровень значимости
df = len(observed_freq) - 1  # Степени свободы
critical_value = stats.chi2.ppf(1 - alpha, df)
p_val = 1 - stats.chi2.cdf(chi2_stat, df)

print(f"p-значение: {p_val}")
print(f"Критическое значение: {critical_value}")

# Принятие решения
if chi2_stat > critical_value:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу")

# Данные на 100% соответствуют распределению Коши. Отклонение от тестового набора минимально и случайно. 
```

```python
# task 5, task 6

chi2_stat, p_val = stats.chisquare(observed_freq, expected_freq)

print(f"Хи-квадрат статистика: {chi2_stat}")
print(f"p-значение: {p_val}")

# Определение критического значения
alpha = 0.05  # Уровень значимости
df = len(observed_freq) - 1  # Степени свободы
critical_value = stats.chi2.ppf(1 - alpha, df)

print(f"Критическое значение: {critical_value}")

# Принятие решения
if chi2_stat > critical_value:
    print("Принимаем альтернативную гипотезу H₁")
else:
    print("Принимаем нулевую гипотезу H₀")

# Данные на 100% соответствуют распределению Коши. Отклонение от тестового набора минимально и случайно. 
```
