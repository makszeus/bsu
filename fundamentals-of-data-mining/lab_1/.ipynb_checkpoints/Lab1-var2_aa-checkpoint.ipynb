```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# TASK 7

df_trees_all = pd.read_csv('avianHabitat.csv')
df_trees_all.head()
```

```python
# task 1
df_trees = df_trees_all[['PW','WHt']]
print(df_trees)
df_trees_filt = df_trees[(df_trees['PW'] != 0) & (df_trees['WHt'] != 0)]
print(df_trees_filt)
```

```python
# task 2
tr_min = df_trees_filt['WHt'].min()
tr_max = df_trees_filt['WHt'].max()
print('min:', tr_min)
print('max:', tr_max)
print('scope:', tr_max - tr_min)
print('mean:', df_trees_filt['WHt'].mean())
print('median:', df_trees_filt['WHt'].median())
print('mode:', df_trees_filt['WHt'].mode()[0])
print('var:', df_trees_filt['WHt'].var())
print('std:', df_trees_filt['WHt'].std())
quart1 = df_trees_filt['WHt'].quantile(0.25)
quart3 = df_trees_filt['WHt'].quantile(0.75)
print('quartile 1:', quart1)
print('quartile 3:', quart3)
print('quartile scope:', quart3 - quart1)
print('skew:', df_trees_filt['WHt'].skew())
print('kurt:', df_trees_filt['WHt'].kurt())
```

```python
# task 5 prep
sorted_tr = df_trees_filt['WHt'].sort_values(ascending=True)
len_tr = len(df_trees_filt['WHt'])
tr_min = df_trees_filt['WHt'].min()
tr_max = df_trees_filt['WHt'].max()
print(sorted_tr)
print('len:', len_tr)
print('min:', tr_min)
print('max:', tr_max)
```

```python
def func_WHt(x):
    if x < tr_min: 
        return 0.
    elif x >= tr_max: 
        return 1.
    else: 
        return ( len(sorted_tr[sorted_tr < x]) / float(len_tr) )

dots = 100
grid = np.linspace(tr_min, tr_max, dots)

vectorized_func_WHt = np.vectorize(func_WHt)
grid_func = vectorized_func_WHt(grid)
```

```python
# task 5 selfmade
plt.plot(grid, grid_func, label='эмпир. функция распределения WHt', color='red')  

plt.xlabel('x')
plt.ylabel('F[WHt](x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# task 5 scipy

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Пример данных
data =  df_trees_filt['WHt']

# Строим функцию плотности распределения
density = stats.gaussian_kde(data)
x = np.linspace(min(data), max(data), 1000)

# Интегрируем функцию плотности для получения функции распределения
cdf = np.cumsum(density(x))
cdf /= cdf[-1]  # Нормализуем CDF, чтобы последний элемент был равен 1

plt.plot(x, cdf)
plt.title("Функция распределения случайной величины (CDF)")
plt.show()
```
