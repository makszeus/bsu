# Задание 1: Использование основных типов графиков.
Используйте набор данных о продажах автомобилей и создайте различные типы графиков, чтобы отобразить различные аспекты данных. 
Включите 
- столбчатые диаграммы, 
- гистограммы, 
- линейные графики, 
- диаграммы размаха
- ящик с усами
- радиальные диаграммы
- scatter plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_cars = pd.read_csv('./data/CAR DETAILS FROM CAR DEKHO.csv')
df_cars_v3 = pd.read_csv('./data/Car details v3.csv')
df_cars_v4 = pd.read_csv('./data/car details v4.csv')
df_cars.head()
```

```python
yearly_sum = df_cars.groupby('year')['selling_price'].sum()
yearly_sum.columns = ['year', 'total_selling_price']

# Building a bar chart.
yearly_sum.plot(kind='bar', x='year', y='total_selling_price', logy = True)
plt.show()
```

```python
df_cars['selling_price'].plot(kind='hist', bins=60, log = True)
plt.show()
```

```python
yearly_sum = df_cars.groupby('year')['selling_price'].sum()
yearly_sum.columns = ['year', 'total_selling_price']
yearly_sum.plot(kind='line', x='year', y='total_selling_price', logy=True)
plt.show()
```

```python
df_cars['selling_price'].plot(kind='box')
plt.yscale('log')
plt.show()
```

```python
df_cars[['selling_price', 'km_driven']].plot(kind='box')
plt.yscale('log')
plt.show()
```

```python
df_cars['fuel'].value_counts().plot(kind='pie')
plt.show()
```

```python
df_cars['fuel'].value_counts()
```

```python
df_cars.plot(kind='scatter', x='km_driven', y='selling_price', loglog = True)
plt.show()
```
