```python
from scipy.special import jv, hermite
from scipy.integrate import quad, solve_ivp
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt
import numpy as np


# 2
def F(a):
    return quad(lambda x: jv(0, a * x) * hermite(5) (x / a), 1, 2)[0]

def task2():
    result = minimize_scalar(F, bounds=(2, 3))
    print(f"Минимальное значение функции на отрезке [2, 3] в точке {result.x}")
    print(result)

# 3
def system(t, y):
    y1, y2, y3 = y
    return [y2, y3, 3 * y3 - 3 * y2 + y1]

def task3():
    interval = [0, 2]
    start_conditions = [0, 0, -1]
    solution = solve_ivp(system, interval, start_conditions, t_eval=np.linspace(0, 2, 100))
    t = solution.t
    y1, y2, y3 = solution.y

    plt.plot(t, y1, label="y(t)")
    plt.plot(t, y2, label="y'(t)")
    plt.plot(t, y3, label="y''(t)")
    plt.legend()
    plt.show()

def main():
    task2()
    task3()

if __name__ == "__main__":
    main()
```
