```python
# Task 12. Variant 9.

import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# Параметры
d = 1.0  # Половина длины отрезка
N = 1000  # Количество точек сетки
x = np.linspace(0, 2*d, N)  # Сетка
h = x[1] - x[0]  # Шаг сетки

# Потенциал V(x) = (x - d)^2
V = (x - d)**2

# Параметры уравнения Шрёдингера
hbar = 1.0  # приведенная постоянная Планка
m = 1.0  # масса частицы

# Шаг по времени для численного метода
dt = 0.001

# Начальное условие: волновая функция в момент времени t=0
Psi = np.zeros(N, dtype=complex)
Psi[0] = 0
Psi[-1] = 0

# Численное решение уравнения Шрёдингера методом явной схемы "вперед"
for t in range(1000):
    # Вычисляем вторую производную волновой функции по x
    d2Psi_dx2 = (Psi[2:] - 2*Psi[1:-1] + Psi[:-2]) / (h**2)
    
    # Обновляем волновую функцию с использованием метода Эйлера
    Psi[1:-1] += (-1j*hbar/(2*m) * d2Psi_dx2 + 1j*V[1:-1]*Psi[1:-1]) * dt

# Нахождение собственных значений и собственных векторов матрицы
eigvals, eigvecs = la.eigh(np.diag(V) - (hbar**2 / (2*m*h**2)) * (np.diag(np.ones(N-1), 1) - 2*np.diag(np.ones(N), 0) + np.diag(np.ones(N-1), -1)))

# Вывод трех минимальных значений энергии и соответствующих волновых функций
for i in range(3):
    print(f'Eigenvalue {i}: {eigvals[i]}')
    plt.plot(x, eigvecs[:, i], label=f'Ψ{i} (E={eigvals[i]:.2f})')

plt.xlabel('x')
plt.ylabel('Ψ(x)')
plt.legend()
plt.title('Eigenfunctions of the Schrödinger equation')
plt.grid()
plt.show()
```
